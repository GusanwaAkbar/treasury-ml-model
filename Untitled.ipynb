```python
# Import library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Data Preparation
data_a = {
    'Amount': [100000, 50000, 70000],
    'Description': ['pembayaran denda angsuran mobil', 'pembayaran stnk mobil', 'pembayaran asuransi']
}

data_b = {
    'Customer': ['a', 'b', 'c', 'd'],
    'PaymentStatus': ['telah membayaran denda angsuran mobil', 'telah membayara pengurusan stnk mobil',
                      'telah membayar asuransi bulanan', 'telah membayar angsuran mobil']
}

df_a = pd.DataFrame(data_a)
df_b = pd.DataFrame(data_b)

# Labeling data in df_a based on df_b
df_a['Verified'] = df_a['Description'].apply(lambda x: any(customer in x for customer in df_b['Customer']))

# Text Preprocessing
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df_a['Description'])

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, df_a['Verified'], test_size=0.2, random_state=42)

# SVM Model
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

# Prediction
y_pred = svm_model.predict(X_test)

# Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

print('\nClassification Report:')
print(classification_report(y_test, y_pred))
```

```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Data Preparation
data_a = {
    'Amount': [100000, 50000, 70000],
    'Description': ['pembayaran denda angsuran mobil', 'pembayaran stnk mobil', 'pembayaran asuransi']
}

data_b = {
    'Customer': ['a', 'b', 'c', 'd'],
    'PaymentStatus': ['telah membayaran denda angsuran mobil', 'telah membayara pengurusan stnk mobil',
                      'telah membayar asuransi bulanan', 'telah membayar angsuran mobil']
}

df_a = pd.DataFrame(data_a)
df_b = pd.DataFrame(data_b)

# Labeling data in df_a based on df_b
df_a['Verified'] = df_a['Description'].apply(lambda x: any(customer in x for customer in df_b['Customer']))
df_a['Label'] = df_a['Verified'].map({True: 'verified', False: 'not_verified'})

# Text Preprocessing
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df_a['Description'])

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, df_a['Label'], test_size=0.2, random_state=42)

# SVM Model
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

# Prediction
y_pred = svm_model.predict(X_test)

# Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

print('\nClassification Report:')
print(classification_report(y_test, y_pred))
```

```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Data Preparation
data_a = {
    'Amount': [100000, 50000, 70000],
    'Description': ['pembayaran denda angsuran mobil', 'pembayaran stnk mobil', 'pembayaran asuransi']
}

data_b = {
    'Customer': ['a', 'b', 'c', 'd'],
    'PaymentStatus': ['telah membayaran denda angsuran mobil', 'telah membayara pengurusan stnk mobil',
                      'telah membayar asuransi bulanan', 'telah membayar angsuran mobil']
}

df_a = pd.DataFrame(data_a)
df_b = pd.DataFrame(data_b)

# Labeling data in df_a based on df_b
df_a['Verified'] = df_a['Description'].apply(lambda x: any(customer in x for customer in df_b['Customer']))

# If there's only one class, add a dummy class to create two classes
if len(df_a['Verified'].unique()) == 1:
    df_a['Verified'] = ~df_a['Verified']  # Invert the values to create two classes

# Text Preprocessing
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df_a['Description'])

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, df_a['Verified'], test_size=0.2, random_state=42)

# SVM Model
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

# Prediction
y_pred = svm_model.predict(X_test)

# Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

print('\nClassification Report:')
print(classification_report(y_test, y_pred))
```
